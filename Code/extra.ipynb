```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

import numpy as np
```

```python
p = np.linspace(0.015, 1, 500)

chi = p**(-1)
```

```python
plt.figure(figsize=(8,8))

plt.loglog(p, chi, color='#e69038')

plt.xlabel(r'$p$', fontsize=18)

plt.ylabel(r'$\chi(p)$', fontsize=18)

plt.savefig('../Results/chi.pdf')
```
